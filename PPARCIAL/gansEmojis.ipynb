# GANs

GANs (*generative adversarial networks*) es una arquitectura de red neuronal propuesta en 2014 por [Ian Goodfellow](https://arxiv.org/pdf/1406.2661.pdf) y otros con el objetivo de obtener modelos capaces de generar datos realistas, principalmente im치genes. Pese a la simplicidad de la idea original, se tard칩 varios a침os en superar varias de las dificultades que presenta su entrenamiento. Hoy en d칤a, sin embargo, se utilizan para obtener resultados espectaculares. Algunos ejemplos son:

- Generaci칩n de caras: https://www.thispersondoesnotexist.com/
- Generaci칩n de anuncios de Airbnb: https://thisrentaldoesnotexist.com/
- Super-resoluci칩n: https://www.youtube.com/watch?v=3RYNThid23g
- Colorizaci칩n: https://github.com/jantic/DeOldify
- Generaci칩n de paisajes: http://nvidia-research-mingyuliu.com/gaugan
- Y muchos otros.

## La arquitectura

La arquitectura b치sica de las *GANs* est치 formada por dos redes neuronales:

- El **generador**: recibe a la entrada valores aleatorios (normalmente obtenidos de una distribuci칩n de tipo gausiana) y da a la salida una imagen. Puedes ver la entrada aleatoria como una representaci칩n latente (o codificaci칩n) de la imagen generada.
- El **discriminador**: recibe a la entrada una imagen (real o generada por el generador) y tiene que decidir si bien la imagen es real o falsa.

![](https://www.researchgate.net/publication/336781462/figure/fig2/AS:817722363494401@1571971612898/Example-of-a-typical-GANs-architecture-for-both-hand-written-digits-recognition-and.png)

Durante el entrenamiento, el generador y el discriminador compiten entre ellos. Por un lado, el generador intenta generar im치genes cada vez m치s realistas para poder enga침ar al discriminador. Por otro lado, el discriminador cada vez ser치 m치s bueno diferenciando im치genes reales de las generadas artificialmente por el generador. Esto implica que no podemos entrenar una *GAN* como hemos entrenado el resto de redes neuronales. El objetivo final ser치 el de obtener un generador tan bueno que nuestro discriminador no sea capaz de diferenciar im치genes reales de falsas. 칄sto impone unas restricciones muy fuertes en el proceso de entrenamiento, haciendo que sea dif칤cil si no lo hacemos de la forma adecuada.

## El entrenamiento

Para poder entrenar una *GAN* de manera efectiva, lo haremos en dos pasos:

1. En una primera fase, entrenamos el discriminador. En cada iteraci칩n, le daremos un *batch* compuesto de im치genes reales obtenidas del dataset e im치genes falsas generadas por el generador. Definiremos un conjunto de etiquetas a 0 para las im치genes falsas y 1 para las reales. Entonces, optimizaremos el discriminador (utilizando una funci칩n de p칠rdida de tipo *binary cross-entropy* mejorando as칤 sus capacidades de distinguir im치genes falsas de reales.
2. En la segunda fase, entrenamos el generador. En cada iteraci칩n, le daremos un *batch* compuesto de ruido aleatorio para que genere im치genes. Estas im치genes son introducidas en el discriminador, cuyas salidas (etiquetas real/falso) con comparadas con un conjunto de etiquetas definidas como reales. Entonces, optimizaremos el generador (utilizando de nuevo la misma funci칩n de p칠rdida) de manera que el generador actualizar치 sus pesos para generar im치genes que enga침en al discriminador.

> 游눠 Como puedes observar, el generador nunca llega a ver las im치genes reales. Sin embargo es capaz de generar im치genes realistas simplemente intentando enga침ar al discriminador. Cu치nto mejor sea el discriminador, mejor ser치 el generador.

```python
from PIL import ImageOps
from torchvision import transforms

class CustomDataset(Dataset):
    def __init__(self, data_dir, transform=None, device='cpu'):
        self.data_dir = data_dir
        self.image_paths = []
        self.labels = []
        self.transform = transform
        self.device = device

        # Obtener la lista de clases (directorios)
        self.classes = os.listdir(data_dir)

        # Recorrer cada clase y sus im치genes
        for label, class_name in enumerate(self.classes):
            class_path = os.path.join(data_dir, class_name)
            if os.path.isdir(class_path):
                # Obtener las rutas de las im치genes en esta clase
                image_paths = [os.path.join(class_path, img) for img in os.listdir(class_path)]
                self.image_paths.extend(image_paths)
                self.labels.extend([label] * len(image_paths))

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        label = self.labels[idx]

        # Cargar la imagen
        image = Image.open(image_path)

        if image.mode == 'RGBA':
            # Rellenar el fondo con un color s칩lido (negro en este caso)
            image = Image.alpha_composite(Image.new('RGBA', image.size, (0, 0, 0, 255)), image)

        # Transformaci칩n adicional: convertir a blanco y negro (escala de grises)
        image = ImageOps.grayscale(image)

        # Redimensionar la imagen a 64x64 p칤xeles
        if self.transform:
            image = self.transform(image)

        # Enviar la imagen y la etiqueta al dispositivo
        image = image.to(self.device)
        label = torch.tensor(label, dtype=torch.long, device=self.device)

        # Transformar la imagen en un tensor 1D
        image = image.view(-1)

        return image, label
```

El objetivo es entrenar una *GAN* que sea capaz de generar im치genes similares a las que tenemos en el dataset a partir de valores aleatorios (ruido). Para ello, como ya hemos visto en las secciones anteriores, necesitamos un **generador** y un **discrimiandor**. Para esta implementaci칩n simple usaremos la misma arquitectura para ambas redes, un Perceptr칩n Multicapa (MLP).

```python
import torch.nn as nn

def block(n_in, n_out):
  return nn.Sequential(
      nn.Linear(n_in, n_out),
      nn.ReLU(inplace=True)
  )

class MLP(nn.Module):
  def __init__(self, input_size, output_size):
    super().__init__()
    self.input_size = input_size
    self.fc1 = block(input_size, 150)
    self.fc2 = block(150, 100)
    self.fc3 = nn.Linear(100, output_size)

  def forward(self, x):
    x = self.fc1(x)
    x = self.fc2(x)
    x = self.fc3(x)
    return x
```

Nuestro MLP est치 formado por 3 capas lineales (cada capa seguida de una activaci칩n `relu`). En funci칩n del n칰mero de entradas y salidas definiremos las diferentes redes. Por ejemplo, el generador recibir치 un vector con 30 valores aleatorios y nos dar치 a la salida un vector de 28 x 28 valores (igual que las im치genes del dataset).

```python
n_in, n_out = 42, 40*40
generator = MLP(n_in, n_out)
```

Obviamente, nuestro generador inicializado genera im치genes aleatorias que en nada se parecen a las reales.

El discriminador, por otro lado, recibir치 a la entrada una imagen (28 x 28 valores) y a la salida nos dar치 una clasificaci칩n binaria (real o falso).

```python
discriminator = MLP(40*40, 1)
output = discriminator(torch.randn(64, 40*40))
output.shape
```

```python
import numpy as np
```

La siguiente funci칩n ser치 la responsable de entrenar ambas redes.

```python
def fit(g, d, dataloader, epochs=30, crit=None):
    g.to(device)
    d.to(device)
    g_optimizer = torch.optim.Adam(g.parameters(), lr=3e-4)
    d_optimizer = torch.optim.Adam(d.parameters(), lr=3e-4)
    crit = nn.BCEWithLogitsLoss() if crit == None else crit
    g_loss, d_loss = [], []
    hist = {'g_loss': [], 'd_loss': []}
    for epoch in range(1, epochs+1):
        for X, y in dataloader:
            #g recibe ruido de 30 por 30 y genera im치genes de 28*28
            g.eval()
            d.train()
            # generamos un batch de im치genes falsas
            noise = torch.randn((X.size(0), g.input_size)).to(device)
            generated_images = g(noise)
            # input del discriminador
            d_input = torch.cat([generated_images, X.view(X.size(0), -1)])
            # ground truth para el discriminador
            d_gt = torch.cat([torch.zeros(X.size(0)), torch.ones(X.size(0))]).view(-1,1).to(device)
            # optimizaci칩n
            d_optimizer.zero_grad()
            d_output = d(d_input)
            d_l = crit(d_output, d_gt)
            d_l.backward()
            d_optimizer.step()
            d_loss.append(d_l.item())
            # entrenamos el generador
            g.train()
            d.eval()
            # generamos un batch de im치genes falsas
            noise = torch.randn((X.size(0), g.input_size)).to(device)
            generated_images = g(noise)
            # salidas del discriminador
            d_output = d(generated_images)
            # ground truth para el generador
            g_gt = torch.ones(X.size(0)).view(-1,1).to(device)
            # optimizaci칩n
            g_optimizer.zero_grad()
            g_l = crit(d_output, g_gt)
            g_l.backward()
            g_optimizer.step()
            g_loss.append(g_l.item())
        # Imprimir los valores de g_loss y d_loss al final de cada 칠poca
        print(f'Epoch {epoch}/{epochs} g_loss {np.mean(g_loss):.5f} d_loss {np.mean(d_loss):.5f}')
        hist['g_loss'].append(np.mean(g_loss))
        hist['d_loss'].append(np.mean(d_loss))
    return hist
```

```python
hist = fit(generator, discriminator, dataloader)
```

```python
import pandas as pd

df = pd.DataFrame(hist)
df.plot(grid=True)
plt.show()
```

Como podemos ver, la *loss* del generador ha ido disminuyendo (lo cual sabemos que es se침al de que el entrenamiento ha ido bien y la red ha aprendido a generar buenas im치genes). Sin embargo, la *loss* del discriminador ha ido aumentando. 쯉ignifica esto que nuestro discriminador es malo, y que tenemos que intentar algo para que este valor tambi칠n disminuya?. Para resolver esta duda tenemos que pensar en c칩mo estamos entrenando ambas redes. Durante el entrenamiento, ambas redes intentan enga침arse mutuamente y van mejorando a la vez en su tarea. Esto implica que en el l칤mite en el que queremos un generador perfecto, el discriminador no ser치 capaz de distinguir im치genes reales de falsas. 쮺u치nto vale la *loss* del discriminador en este caso? La respuesta es 0.5 (cada vez que el discriminador recibe una imagen le asigna la etiqueta real o falso con un 50% de probabilidad ya que no es capaz de distinguirlas). As칤 pues, en el contexto de las *GANs*, que la *loss* del discriminador aumente es realmente una buena se침al de que todo est치 funcionando como esperamos.

## DCGANs

Podemos usar *DCGANs* (*deep convolutional GANs*) para obtener mejores generadores utilizando redes convolucionales. En este caso, necesitaremos arquitecturas diferentes para generador y discriminador.

![](https://gluon.mxnet.io/_images/dcgan.png)

El generador recibir치 un vector de 100 valores aleatorios y despu칠s aplicaremos varias capas de convoluciones transpuestas (que aumentar치n el tama침o de los mapas de caracter칤sitcas, como hicimos en nuestras redes para [segmentaci칩n](https://sensioai.com/blog/50_cv_segmentacion)) hasta obtener la imagen generada.

```python
class Generator(nn.Module):
  def __init__(self):
    super().__init__()
    self.input_size = 100
    self.inp = nn.Sequential(
        nn.Linear(self.input_size, 10*10*128),
        nn.BatchNorm1d(10*10*128),
    )
    self.main = nn.Sequential(
        nn.ConvTranspose2d(128, 64, 4, stride=2, padding=1, bias=False),
        nn.BatchNorm2d(64),
        nn.ReLU(True),
        nn.ConvTranspose2d(64, 1, 4, stride=2, padding=1, bias=False),
        nn.Tanh()
    )

  def forward(self, x):
    x = self.inp(x)
    x = x.view(-1, 128, 10, 10)
    x = self.main(x)
    x = x.view(x.size(0), 40*40)
    return x
```

```python
generator = Generator()
output = generator(torch.randn(64, 100))
output.shape
```

Al final del generador usamos una activaci칩n `tanh`, que dar치 valores entre -1 y 1. Por este motivo tenemos que re-normalizar nuestras im치genes en el dataset.

```python
from PIL import ImageOps
from torchvision import transforms

class CustomDataset(Dataset):
    def __init__(self, data_dir, transform=None, device='cpu'):
        self.data_dir = data_dir
        self.image_paths = []
        self.labels = []
        self.transform = transform
        self.device = device

        # Obtener la lista de clases (directorios)
        self.classes = os.listdir(data_dir)

        # Recorrer cada clase y sus im치genes
        for label, class_name in enumerate(self.classes):
            class_path = os.path.join(data_dir, class_name)
            if os.path.isdir(class_path):
                # Obtener las rutas de las im치genes en esta clase
                image_paths = [os.path.join(class_path, img) for img in os.listdir(class_path)]
                self.image_paths.extend(image_paths)
                self.labels.extend([label] * len(image_paths))

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        label = self.labels[idx]

        # Cargar la imagen
        image = Image.open(image_path)

        if image.mode == 'RGBA':
            # Rellenar el fondo con un color s칩lido (negro en este caso)
            image = Image.alpha_composite(Image.new('RGBA', image.size, (0, 0, 0, 255)), image)

        # Transformaci칩n adicional: convertir a blanco y negro (escala de grises)
        image = ImageOps.grayscale(image)

        # Redimensionar la imagen a 64x64 p칤xeles
        if self.transform:
            image = self.transform(image)

        # image = (image * 2.0) - 1.0

        # Enviar la imagen y la etiqueta al dispositivo
        image = image.to(self.device)
        label = torch.tensor(label, dtype=torch.long, device=self.device)

        # Transformar la imagen en un tensor 1D
        image = image.view(-1)

        return image, label
```

```python
# Define las transformaciones que deseas aplicar a las im치genes (puedes personalizarlas seg칰n tus necesidades)
transform = transforms.Compose([transforms.Resize((40, 40)),
                                transforms.ToTensor()])

# Especifica la ruta de tu conjunto de datos
data_dir = '/content/drive/MyDrive/IA22023/DATASETS/image/image'

# Crea una instancia del conjunto de datos personalizado
train = CustomDataset(data_dir, transform=transform, device=device)

# Obtiene la longitud del conjunto de datos
dataset_length = len(train)
```

En cuanto al discriminador, utilizaremos una `CNN` t칤pica como las que conocemos cuando hacemos clasificaci칩n de imagen.

```python
class Discriminator(nn.Module):
  def __init__(self):
    super().__init__()
    self.main = nn.Sequential(
        nn.Conv2d(1, 64, 4, stride=2, padding=1, bias=False),
        nn.BatchNorm2d(64),
        nn.ReLU(True),
        nn.Conv2d(64, 128, 4, stride=2, padding=1, bias=False),
        nn.BatchNorm2d(128),
        nn.ReLU(True)
    )
    self.out = nn.Sequential(
        nn.Linear(128*10*10, 1),
        nn.Sigmoid()
    )

  def forward(self, x):
    # esperamos vectores a la entrada de 64*64
    x = x.view(x.size(0), 1, 40, 40)
    x = self.main(x)
    x = x.view(x.size(0), -1)
    x = self.out(x)
    return x
```

```python
discriminator = Discriminator()
output = discriminator(torch.randn(64, 40*40))
output.shape
```

## Resumen

En este post hemos aprendido a implementar *GANs*, redes neuronales capaces de generar datos similares a los usados en el entrenamiento, en este caso im치genes. Este tipo de arquitecturas est치n formadas por dos redes neuronales que compiten entre s칤 durante el entrenamiento: el generador y el discriminador. El generador se encarga de generar im치genes falsas a partir de un vector de valores aleatorios, que se puede interpretar como una versi칩n comprimida de la imagen, mientras que el discriminador se encarga de distinguir entre im치genes reales, obtenidas del dataset, y falsas, generadas por el generador. Una vez entrenado el modelo, descartamos el discriminador y nos quedamos con el generador que ser치 capaz de generar im치genes realistas, similares a la usadas en el dataset. Este tipo de modelos se utilizan para generar caras de gente que no existe, colorear im치genes en blanco y negro, generar im치genes realistas a partir de m치scaras de segmentaci칩n, incluso los famosos *deep fakes*.
