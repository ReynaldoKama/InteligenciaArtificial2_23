```python
import cv2
from IPython.display import display, Image
from pathlib import Path
from ultralytics import YOLO  # AsegÃºrate de importar la clase YOLO desde tu implementaciÃ³n

# Nombre de la imagen
# img_name = 'img_3_jpg.rf.0a7f3cc38f4dd6b117b20b3e05eb826f'

# Ruta de la imagen
img_path = f"/content/drive/MyDrive/YOLO8/20231109_111015.jpg"

# Mostrar la imagen en una ventana
img = cv2.imread(img_path)
display(Image(filename=img_path, width=300))

# Inicializar el modelo YOLO
model = YOLO("/content/drive/MyDrive/YOLO8/dataset/runs/segment/train/detect/train/weights/best.pt")

# Realizar la predicciÃ³n
pred = model.predict(img)[0]

# Dibujar las cajas delimitadoras y etiquetas en la imagen
img_pred = pred.plot()
cv2.imwrite(f"20231109_111015.jpg", img_pred)

# Mostrar la imagen con las cajas delimitadoras y etiquetas
display(Image(filename=f"20231109_111015.jpg", width=300))
```
